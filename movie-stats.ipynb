```python
import pandas as pd 

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('C:/Users/bourb/users.dat', sep='::', header=None, names=unames)
#users = pd.read_table('users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('C:/Users/bourb/ratings.dat', sep='::', header=None, names=rnames)
#ratings = pd.read_table('ratings.dat', sep='::', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('C:/Users/bourb/movies.dat', sep='::', header=None, names=mnames)
#movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)

data[:10]
```

```python
data.loc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')

mean_ratings[:10]
```

```python
ratings_by_title = data.groupby('title').size()

ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]

active_titles[:10]
```

```python
meaner_ratings = mean_ratings.loc[active_titles]

meaner_ratings
```

```python
top_female_ratings = meaner_ratings.sort_values(by='F', ascending=False)

top_female_ratings[:10]
```

```python
top_male_ratings = meaner_ratings.sort_values(by='M', ascending=False)

top_male_ratings[:10]
```

```python
meaner_ratings['diff'] = meaner_ratings['M'] - meaner_ratings['F']

sorted_by_diff = meaner_ratings.sort_values(by='diff')

sorted_by_diff[:25]
```

```python
sorted_by_diff[::-1][:25]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()

rating_std_by_title = rating_std_by_title.loc[active_titles]

rating_std_by_title.sort_values(ascending=False)[:10]
```
